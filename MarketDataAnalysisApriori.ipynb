```python
import pandas as pd
df = pd.read_csv("/content/Market_Basket_Optimisation.csv", header = None)
df
```

```python
import numpy as np
import matplotlib.pyplot as ply
from apyori import apriori
```

```python
df.fillna(0, inplace = True) # Bosluqlari 0 ile doldurdu.
```

```python
transactions = []    #
for i in range(0, len(df)):  # 0 dan basla butun setirleri gez.
  transactions.append([str(df.values[i,j]) for j in range (0,20) if str(df.values[i,j]) != '0'])  # liste birlesdirir, j ni 0 dan 20 ye qeder qoyur. get eger 20 ci setirde
```

```python
transactions
```

```python
transactions[1]
```

```python
rules = apriori(transactions, min_support = 0.001, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
rules
```

```python
results = list(rules)
results
```

```python
df_results = pd.DataFrame(results)
df_results
```

```python
support = df_results.support #1000 dene alis verisde en azi 3 dene tekrarlanani gostersin.
```

LHS: The expected or average highest order statistic for a sample of size

RHS: The probability density function of the expected highest order statistic.


```python
first_values = []
second_values = []
third_values = []
fourth_value = []
```

```python
for i in range(df_results.shape[0]):
  single_list = df_results['ordered_statistics'][i][0]
  first_values.append(list(single_list[0]))
  second_values.append(list(single_list[1]))
  third_values.append(single_list[2])
  fourth_value.append(single_list[3])
```

```python
list(single_list)
```

```python
lhs = pd.DataFrame(first_values)
rhs = pd.DataFrame(second_values)

confidence = pd.DataFrame(third_values, columns = ['Confidence'])

lift = pd.DataFrame(fourth_value, columns = ['lift'])
```

```python
df_final = pd.concat([lhs, rhs, support, confidence, lift], axis = 1)
df_final
```

```python
df_final.fillna(value = '', inplace = True)
df_final.head()
```

```python
df_final.columns = ['lhs_1', 'lhs_2', 'lhs_3', 'lhs_4', 'rhs_1', 'rhs_2', 'rhs_3', 'support', 'confidence', 'lift']
df_final.head()
```

```python
df_final['lhs_1'] = df_final['lhs_1'] + str(",") + df_final['lhs_2']

df_final['rhs_1'] = df_final['rhs_2'] + str(" , ") + df_final['rhs_2'] + str(",") + df_final['rhs_3']
```

```python
df_final.drop(columns = [1,2,3], inplace = True)
df_final
```

```python
df_final.sort_values('lift', ascending = False)
```
