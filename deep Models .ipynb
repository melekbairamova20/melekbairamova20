```python
def downsampling(inputs, n_filter):
  x = tf.keras.layers.Conv2D(n_filter, kernel_size = (3, 3), activation = 'relu', kernel_initializer = 'he_normal')(inputs)
  concat = tf.keras.layers.Conv2D(n_filter, kernel_size = (3, 3), activation = 'relu', kernel_initializer = 'he_normal')(x)
  max_pool = tf.keras.layers.MaxPool2D(pool_size = (2, 2))(concat)

  return concat, max_pool
```

```python
import tensorflow as tf
n_classes = 4
inputs = tf.keras.layers.Input(shape = (572,572,3))
concat_1, max_pool1 = downsampling(inputs , 64)
concat_2, max_pool2 = downsampling(max_pool1, 128)
concat_3, max_pool3 = downsampling(max_pool2, 256)
concat_4, max_pool4 = downsampling(max_pool3, 512)



conv_0 = tf.keras.layers.Conv2D(512, kernel_size = (3,3),activation = "relu" , kernel_initializer = "he_normal")(max_pool4)
conv_1 = tf.keras.layers.Conv2D(512, kernel_size = (3,3),activation = "relu" , kernel_initializer = "he_normal")(conv_0)

upsample_0 = tf.keras.layers.Conv2DTranspose(filters = 512, kernel_size = (2, 2), activation = 'relu', kernel_initializer = 'he_normal')(conv_1)

concat_4_reshaped = tf.keras.layers.Lambda(
    lambda x:tf.image.resize(x[0], size = (x[1].shape[1], x[1].shape[2]))

)([concat_4, upsample_0])
concatenate_0 = tf.keras.layers.Concatenate(axis = -1)([concat_4_reshaped, upsample_0])

conv_2 = tf.keras.layers.Conv2D(512, kernel_size = (3, 3), activation = 'relu', kernel_initializer = 'he_normal')(concatenate_0)
upsample_1 = tf.keras.layers.Conv2D(512, kernel_size = (3, 3), activation = 'relu', kernel_initializer = 'he_normal')(conv_2)

concat_3_reshaped = tf.keras.layers.Lambda(
    lambda x: tf.image.resize(x[0], size = (x[1].shape[1], x[1].shape[2]))

)([concat_3, upsample_1])
concatenate_1 = tf.keras.layers.Concatenate(axis = -1)([concat_3_reshaped, upsample_1])

conv_3 = tf.keras.layers.Conv2D(256, kernel_size = (3, 3), activation = 'relu', kernel_initializer = 'he_normal')(concatenate_1)
upsample_2 = tf.keras.layers.Conv2D(256, kernel_size = (3, 3), activation = 'relu', kernel_initializer = 'he_normal')(conv_3)

concat_2_reshaped = tf.keras.layers.Lambda(
    lambda x: tf.image.resize(x[0], size = (x[1].shape[1], x[1].shape[2]))
)([concat_2, upsample_2])
concatenate_2 = tf.keras.layers.Concatenate(axis = -1)([concat_2_reshaped, upsample_2])

conv_4 = tf.keras.layers.Conv2D(198, kernel_size = (3,3), activation = 'relu', kernel_initializer = 'he_normal')(concatenate_2)
upsample_3 = tf.keras.layers.Conv2D(198, kernel_size = (3,3), activation = 'relu', kernel_initializer = 'he_normal')(conv_4)

concat_1_reshaped = tf.keras.layers.Lambda(
    lambda x: tf.image.resize(x[0], size = (x[1].shape[1], x[1].shape[2]))

)([concat_1, upsample_3])
concatenate_3 = tf.keras.layers.Concatenate(axis = -1)([concat_1_reshaped, upsample_3])

layer_0, _ = downsampling(concatenate_3, 64)
flattened = tf.keras.layers.Flatten()(layer_0)
outputs = tf.keras.layers.Dense(n_classes, activation = 'softmax')(flattened)
```

```python
unet = tf.keras.Model(inputs = inputs, outputs = outputs)
unet.summary()
```

```python
tf.keras.utils.plot_model(unet, show_shapes = True)
```

```python
unet.compile(loss = 'binary_crossentropy', optimizer = tf.keras.optimizers.Adam(), metrics = ['accuracy'])
```
