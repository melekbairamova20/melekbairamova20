```python
import numpy as np
np.random.seed(42)
m = 100
X =  np.random.rand(m,1)
y = 2 * X
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,6))
plt.scatter(X,y)
plt.xlabel('$X$')
plt.ylabel('$Y$', rotation = 0);

plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
from sklearn.preprocessing import add_dummy_feature
new_X = np.array([[0], [2]])

y_predict = lin_reg.predict(np.array([0, 1]).reshape(-1,1))

```

```python
plt.figure(figsize = (10,6))
plt.plot(X,y, 'b.')
plt.plot(y_predict, 'r-')
plt.xlabel('$X$')
plt.ylabel('$Y$', rotation = 0)
plt.grid()
plt.show()
```

```python

pipeline = make_pipeline(PolynomialFeatures(degree = 2), LinearRegression())

pipeline.fit(X.reshape(-1,1), y)
X_new = np.linspace(1, 100, 100).reshape(-1,1)
preds = pipeline.predict(X_new)
plt.plot(X, y, 'b.')
plt.plot(X_new, preds, 'r-')
```
