```python
cars = {"BMW" : 100, "Bugatti" : 200, "Tesla" : 300}

plt.bar(cars.keys(), cars.values())
plt.xlabel("Car names")  # x oxuna car names yazir
plt.ylabel("Car prices")  # y oxuna car prices yazir
plt.title("Car prices");  #basliga ad qoyur
```

```python
plt.barh(list(cars.keys()), list(cars.values()));  # barh -> datani list formatinda isteyir.
```

```python
votes = [35,40,35]
labels = ['Phyton', 'C++', 'Javascript']
plt.pie(votes,labels = labels, explode = [0, 0, 0.2], autopct = '%.2f%%'); # explode birini ayirib cixarir, autopct olan hisse faizle qaytarir.(noqteden sonraki 2  noqteden sonra nece reqem geleceyini  gosterir)
```
